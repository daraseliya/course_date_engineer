# 1. Построение распределений

```python
import numpy as np
import matplotlib.pyplot as plt

for s in range(1,5):
  x = np.arange(0, 10.01, 0.01)
  plt.plot(x, (x/s**2)*np.exp((-x**2)/(2*s**2)), label =  s )

plt.legend(fontsize=12, loc=1) 
plt.title(r'Плотность вероятности для распределение Рэлея', fontsize=12) 
plt.xlabel(r'$x$', fontsize=12) 
plt.ylabel(r'Плотность, $f (x,\sigma)$ ', fontsize=12) 
plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np

for s in range(4,5):
  x = np.arange(0, 10.01, 2/3)
  plt.bar(x,  (x/s**2)*np.exp((-x**2)/(2*s**2)), label = 'Гистограмма вероятности')
  plt.plot(x,  (x/s**2)*np.exp((-x**2)/(2*s**2)), color = 'red',  label = 'Плотность вероятности')

plt.legend(fontsize=12, loc=1) 
plt.title(r'Плотность вероятности для распределение Рэлея, $\sigma$=4', fontsize=12)
plt.xlabel(r'$x$', fontsize=12) 
plt.ylabel(r'Плотность, $f (x,\sigma)$ ', fontsize=12) 
plt.show()
```

# 3. Анализ зависимостей

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = [] # возраст
sex = [] #пол
bmi = [] #индекc массы тела
children = [] #кол-во детей
smoker = [] #курит
region = [] #регион
charges = [] #то сколько человек тратит на страховки
```

```python
ls
```

```python
with open("./dataset.txt", mode="r") as file:
  for human in file.readlines():
    age.append(int(human.split(",")[0])) 

    if human.split(",")[1] == "male":
      sex.append(1)
    else:
      sex.append(0)

    bmi.append(float(human.split(",")[2]))

    children.append(float(human.split(",")[3]))

    if human.split(",")[4] == "yes":
      smoker.append(1)
    else:
      smoker.append(0)

    if human.split(",")[5] == "northwest":
      region.append(0)
    elif human.split(",")[5] == "northeast":
      region.append(1)
    elif human.split(",")[5] == "southwest":
      region.append(2)
    elif human.split(",")[5] == "southeast":
      region.append(3)
    else:
      region.append(human.split(",")[5])

    charges.append(float(human.split(",")[6]))
```

```python
age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
charges = np.array(charges)
```
