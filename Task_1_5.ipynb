##7. Сколько должностей в наборе данных?

```python
df2.groupby("job_title").job_title.count()
```

```python
job_names = list(df2.job_title)
```

```python
#for i in range(len(job_names)):
#  print(job_names[i])
```

```python
#job_names_dict = {}
```

```python
job_names_dict = {}
j_nam_count = 0
for i in job_names:
  if i in job_names_dict:
    job_names_dict[i] +=1
  else:
    job_names_dict[i]=1
    j_nam_count +=1
```

```python
job_names_dict
```

```python
j_nam_count
```

```python
# Ответ: кол-во должностей всего 50
```

##8 Какие 10 наиболее часто встречающихся должностей?

```python
for key,value in job_names_dict.items():
  if value>7:
    print(key,value)
```
